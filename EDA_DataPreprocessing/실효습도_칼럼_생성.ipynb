```python
import pandas as pd
```

```python
상대습도17 = pd.read_csv("C:/Users/LG/Desktop/2017 상대습도.csv", encoding = 'cp949')
상대습도17=상대습도17[['평균 상대습도(%)','일시']]
상대습도18 = pd.read_csv("C:/Users/LG/Desktop/2018 상대습도.csv", encoding = 'cp949')
상대습도18=상대습도18[['평균 상대습도(%)','일시']]
상대습도19 = pd.read_csv("C:/Users/LG/Desktop/2019 상대습도.csv", encoding = 'cp949')
상대습도19=상대습도19[['평균 상대습도(%)','일시']]
상대습도20 = pd.read_csv("C:/Users/LG/Desktop/2020 상대습도.csv", encoding = 'cp949')
상대습도20=상대습도20[['평균 상대습도(%)','일시']]
상대습도21 = pd.read_csv("C:/Users/LG/Desktop/2021 상대습도.csv", encoding = 'cp949')
상대습도21=상대습도21[['평균 상대습도(%)','일시']]
```
